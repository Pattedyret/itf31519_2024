# A Machine Learning solution to assignment: When do we play tennis?

The following code demonstrates how to use a decision tree to solve the assignment posted in Canvas. 

```python
# Read and print the dataset
data_frame = pd.read_csv('./dataset.csv').astype('category')
print(data_frame)

# Divide into input and output
feature_frame = data_frame[['Outlook','Temperature','Humidity','Wind']]
target_series = data_frame['Play?']

# Convert string data to encoded categories
categorical_feature_frame = pd.get_dummies(feature_frame)

# Print encoded result
print(categorical_feature_frame)
print(target_series)
```

```python
# Import decision tree functionality
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Create decision tree and fit it to the data
classifier = DecisionTreeClassifier()
classifier.fit(categorical_feature_frame, target_series)

# Show the tree 
plt.figure(figsize=(10,10))
plot_tree(classifier)

# Test the accuracy of the classifier on the training data
accuracy = classifier.score(categorical_feature_frame, target_series)
print('The accuracy of the classifier on training data is ' + str(accuracy))
```
